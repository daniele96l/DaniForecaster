```python
import time
from datetime import date

import pandas as pd
import requests
from pandas.tseries.offsets import MonthEnd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

BASE_URL = "https://api.marketstack.com/v1/"
start_date = "2003-12-29"
today = date.today().isoformat()
print(f"Fetching data from {start_date} to {today}")


def daily_to_monthly(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values(["Ticker", "Date"]).set_index("Date")
    monthly = (
        df.groupby(["Ticker", pd.Grouper(freq=MonthEnd())])
        .agg(Open=("Open", "first"))
        .reset_index()
    )
    return monthly.rename(columns={"Date": "MonthEnd"})


def add_next_month_direction(monthly: pd.DataFrame) -> pd.DataFrame:
    out = monthly.sort_values(["Ticker", "MonthEnd"]).copy()
    nxt = out.groupby("Ticker")["Open"].shift(-1)
    up = nxt > out["Open"]
    down = nxt < out["Open"]
    out["next_month_up"] = pd.Series(pd.NA, index=out.index, dtype="Int64")
    out.loc[up, "next_month_up"] = 1
    out.loc[down, "next_month_up"] = 0
    out.loc[nxt.isna(), "next_month_up"] = pd.NA
    return out


def add_open_price_features(monthly: pd.DataFrame) -> pd.DataFrame:
    out = monthly.sort_values(["Ticker", "MonthEnd"]).copy()
    grouped_open = out.groupby("Ticker")["Open"]

    out["open_pct_change_1m"] = grouped_open.pct_change(1)
    out["open_pct_change_3m"] = grouped_open.pct_change(3)

    out["open_pct_change_1m_lag1"] = out.groupby("Ticker")["open_pct_change_1m"].shift(1)
    out["open_pct_change_1m_lag2"] = out.groupby("Ticker")["open_pct_change_1m"].shift(2)
    out["open_pct_change_1m_lag3"] = out.groupby("Ticker")["open_pct_change_1m"].shift(3)

    out["open_ret_roll_mean_3"] = out.groupby("Ticker")["open_pct_change_1m"].transform(
        lambda s: s.rolling(window=3).mean()
    )
    out["open_ret_roll_std_3"] = out.groupby("Ticker")["open_pct_change_1m"].transform(
        lambda s: s.rolling(window=3).std()
    )
    out["open_ret_roll_mean_6"] = out.groupby("Ticker")["open_pct_change_1m"].transform(
        lambda s: s.rolling(window=6).mean()
    )
    out["open_ret_roll_std_6"] = out.groupby("Ticker")["open_pct_change_1m"].transform(
        lambda s: s.rolling(window=6).std()
    )

    out["open_vs_roll3"] = (out["Open"] / grouped_open.transform(lambda s: s.rolling(window=3).mean())) - 1
    return out
```

# Data gathering

Pull end-of-day (EOD) prices from the Marketstack API for every ticker that ever appeared in the Nasdaq-100 reconstruction lists below. The notebook paginates through the full history, then saves a single consolidated table to `df_nasdaq_eod.csv`.


```python
all_stock_data = []
# Paginate until a short page; do not use pagination.total (often capped at 1000 → one page only).
PAGE = 1000

for ticker in full_ticker_list:
    offset = 0
    while True:
        params = {
            'access_key': API_KEY_MARKETSTACK,
            'symbols': ticker,
            'date_from': start_date,
            'date_to': today,
            'limit': PAGE,
            'offset': offset,
        }

        try:
            response = requests.get(f"{BASE_URL}eod", params=params)
            response.raise_for_status()
            data = response.json()

            rows = (data or {}).get('data') or []
            if not rows:
                if offset == 0:
                    print(f"No data for {ticker}")
                break

            print(f"Fetched {len(rows)} rows for {ticker} (offset {offset})")

            for entry in rows:
                all_stock_data.append({
                    'Date': entry['date'].split('T')[0],
                    'Ticker': entry['symbol'],
                    'Open': entry['open'],
                })

            if len(rows) < PAGE:
                break

            offset += len(rows)
            time.sleep(0.1)

        except Exception as e:
            print(f"Error for {ticker} at offset {offset}: {e}")
            break

# Create DataFrame
if all_stock_data:
    df_nasdaq_eod = pd.DataFrame(all_stock_data)
    df_nasdaq_eod['Date'] = pd.to_datetime(df_nasdaq_eod['Date'])
    df_nasdaq_eod = df_nasdaq_eod.sort_values(['Ticker', 'Date'])
    print(f"\nSuccessfully fetched {len(df_nasdaq_eod)} total rows across all tickers.")
else:
    df_nasdaq_eod = pd.DataFrame(columns=['Date', 'Ticker', 'Open'])
    print("No data was collected.")
```

```python
import pandas as pd

try:
    df_nasdaq_monthly = pd.read_csv("df_nasdaq_monthly_features.csv")
except FileNotFoundError:
    df_nasdaq_monthly = pd.read_csv("df_nasdaq_monthly.csv")
    df_nasdaq_monthly = add_open_price_features(df_nasdaq_monthly)

df_nasdaq_monthly["MonthEnd"] = pd.to_datetime(df_nasdaq_monthly["MonthEnd"])
```

```python
feature_cols = [
    "open_pct_change_1m",
    "open_pct_change_3m",
    "open_pct_change_1m_lag1",
    "open_pct_change_1m_lag2",
    "open_pct_change_1m_lag3",
    "open_ret_roll_mean_3",
    "open_ret_roll_std_3",
    "open_ret_roll_mean_6",
    "open_ret_roll_std_6",
    "open_vs_roll3",
]
if not set(feature_cols).issubset(df_nasdaq_monthly.columns):
    df_nasdaq_monthly = add_open_price_features(df_nasdaq_monthly)

model_df = df_nasdaq_monthly.dropna(subset=["next_month_up", *feature_cols]).copy()
model_df["next_month_up"] = model_df["next_month_up"].astype(int)

unique_months = sorted(model_df["MonthEnd"].unique())
split_idx = int(len(unique_months) * 0.8)
split_month = unique_months[split_idx]

train_df = model_df[model_df["MonthEnd"] < split_month].copy()
test_df = model_df[model_df["MonthEnd"] >= split_month].copy()

X_train = train_df[feature_cols]
y_train = train_df["next_month_up"]
X_test = test_df[feature_cols]
y_test = test_df["next_month_up"]

class_counts = y_train.value_counts(normalize=True)
minority_share = class_counts.min()
class_weight = "balanced" if minority_share < 0.4 else None

model = RandomForestClassifier(
    n_estimators=300,
    max_depth=8,
    random_state=42,
    class_weight=class_weight,
)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Train rows:", len(train_df), "Test rows:", len(test_df))
print("Train month range:", train_df["MonthEnd"].min(), "to", train_df["MonthEnd"].max())
print("Test month range:", test_df["MonthEnd"].min(), "to", test_df["MonthEnd"].max())
print("Accuracy:", round(accuracy_score(y_test, y_pred), 4))
print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))
print("Classification report:")
print(classification_report(y_test, y_pred, digits=4))

feature_importance = (
    pd.DataFrame({"feature": feature_cols, "importance": model.feature_importances_})
    .sort_values("importance", ascending=False)
    .reset_index(drop=True)
)
print("Feature importance:")
print(feature_importance)

predictions_df = test_df[["Ticker", "MonthEnd", "next_month_up"]].copy()
predictions_df["pred_next_month_up"] = y_pred
predictions_df.to_csv("next_month_up_predictions.csv", index=False)
```
