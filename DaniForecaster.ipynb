```python
import time
from datetime import date

import pandas as pd
import requests
from pandas.tseries.offsets import MonthEnd

BASE_URL = "https://api.marketstack.com/v1/"
start_date = "2003-12-29"
today = date.today().isoformat()
print(f"Fetching data from {start_date} to {today}")


def daily_to_monthly(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values(["Ticker", "Date"]).set_index("Date")
    monthly = (
        df.groupby(["Ticker", pd.Grouper(freq=MonthEnd())])
        .agg(Open=("Open", "first"))
        .reset_index()
    )
    return monthly.rename(columns={"Date": "MonthEnd"})


def add_next_month_direction(monthly: pd.DataFrame) -> pd.DataFrame:
    out = monthly.sort_values(["Ticker", "MonthEnd"]).copy()
    nxt = out.groupby("Ticker")["Open"].shift(-1)
    up = nxt > out["Open"]
    down = nxt < out["Open"]
    out["next_month_up"] = pd.Series(pd.NA, index=out.index, dtype="Int64")
    out.loc[up, "next_month_up"] = 1
    out.loc[down, "next_month_up"] = 0
    out.loc[nxt.isna(), "next_month_up"] = pd.NA
    return out
```

# Data gathering

Pull end-of-day (EOD) prices from the Marketstack API for every ticker that ever appeared in the Nasdaq-100 reconstruction lists below. The notebook paginates through the full history, then saves a single consolidated table to `df_nasdaq_eod.csv`.


```python
all_stock_data = []
# Paginate until a short page; do not use pagination.total (often capped at 1000 → one page only).
PAGE = 1000

for ticker in full_ticker_list:
    offset = 0
    while True:
        params = {
            'access_key': API_KEY_MARKETSTACK,
            'symbols': ticker,
            'date_from': start_date,
            'date_to': today,
            'limit': PAGE,
            'offset': offset,
        }

        try:
            response = requests.get(f"{BASE_URL}eod", params=params)
            response.raise_for_status()
            data = response.json()

            rows = (data or {}).get('data') or []
            if not rows:
                if offset == 0:
                    print(f"No data for {ticker}")
                break

            print(f"Fetched {len(rows)} rows for {ticker} (offset {offset})")

            for entry in rows:
                all_stock_data.append({
                    'Date': entry['date'].split('T')[0],
                    'Ticker': entry['symbol'],
                    'Open': entry['open'],
                })

            if len(rows) < PAGE:
                break

            offset += len(rows)
            time.sleep(0.1)

        except Exception as e:
            print(f"Error for {ticker} at offset {offset}: {e}")
            break

# Create DataFrame
if all_stock_data:
    df_nasdaq_eod = pd.DataFrame(all_stock_data)
    df_nasdaq_eod['Date'] = pd.to_datetime(df_nasdaq_eod['Date'])
    df_nasdaq_eod = df_nasdaq_eod.sort_values(['Ticker', 'Date'])
    print(f"\nSuccessfully fetched {len(df_nasdaq_eod)} total rows across all tickers.")
else:
    df_nasdaq_eod = pd.DataFrame(columns=['Date', 'Ticker', 'Open'])
    print("No data was collected.")
```
